Diseña un algoritmo para resolver el problema por fuerza bruta

### Fuerza bruta. 

```python
from itertools import permutations 

DIGITOS = ['1', '2', '3', '4', '5', '6', '7', '8', '9'] # lista de digitos
OPERADORES = ['+', '-', '*', '/'] #lista de operadores

def construir_expresion(numeros, operadores): 
# La función recibe una tupla con 5 digitos y otra tupla de 4 operadores y arma una expresión del tipo: n1 op1 n2 op2 n3 op3 n4 op4 n5
    expresion = ""
    for i in range(len(operadores)): # Recorre los índices de los operadores (0 a 3)
        expresion += numeros[i] + operadores[i]
    expresion += numeros[-1] # Cuando sale del bucle for agrega el último digito. 
    return expresion

def es_entero(valor, tolerancia=1e-9):  # Verifica si el valor que devuelve la expresión es entero, comparando con un grado de tolerancia. 
    return abs(valor - round(valor)) < tolerancia

resultados = set() # Conjunto que guarda los valores enteros resultantes de la expresión evitando duplicados
ejemplos = {} #  guarda una expresión para cada item de resultados.

for nums in permutations(DIGITOS, 5): # Genera todas las permutaciones posibles de 5 dígitos distintos tomados de los 9 disponibles (15120)
    for ops in permutations(OPERADORES, 4): # Genera todas las permutaciones posibles de los 4 operadores (24). 
        expr = construir_expresion(nums, ops) # llama a la función definida previamente con las tuplas nums, ops como parámetros

        try:
            valor = eval(expr) # Evalúa la expresión  construida a partir de los dígitos y operadores mediante eval
            if es_entero(valor):  # Llama a la función que verifica si el valor es un número aproximadamente entero. 
                valor_int = int(round(valor)) # Convierte mediante la función round, corrigiendo posibles errores de precisión"
                resultados.add(valor_int) # El valor se agrega al conjunto de resultados

                if valor_int not in ejemplos:
                    ejemplos[valor_int] = expr

        except ZeroDivisionError: # Se deja esta excepción aunque con el caso planteado es poco probable que se alcance la situación. 
            pass

print("Cantidad de enteros:", len(resultados))
print("Mínimo:", min(resultados))
print("Máximo:", max(resultados))

faltantes = [x for x in range(min(resultados), max(resultados) + 1) if x not in resultados]

print("¿Están todos los enteros entre mínimo y máximo?", len(faltantes) == 0)
print("Algunos faltantes:", faltantes[:10])

print("\nEjemplos:")
for v in sorted(resultados)[:10]:
    print(v, "=>", ejemplos[v])
```

```python
DIGITOS = ['1','2','3','4','5','6','7','8','9']
OPERADORES = ['+','-','*','/']

enteros = set() # Conjunto que guarda los valores enteros resultantes de la expresión evitando duplicados
resultados = {} # Guarda una expresión por cada item de resultado. 

def es_prometedora(solucion, etapa): # Función que verifica que la solución parcial no tenga digitos repetidos ni operadores en posicioens pares
    digitos_usados = [] 
    operadores_usados = [] 

    for i in range(etapa + 1): # Se recorre la solución hasta la etapa actual
        if i % 2 == 0:
            if solucion[i] in digitos_usados:
                return False  # no prometedora - dígito repetido
            digitos_usados.append(solucion[i])
        else:
            if solucion[i] in operadores_usados:
                return False  # no prometedora - operador repetido
            operadores_usados.append(solucion[i])

    return True  # es prometedora
       
def expresion_valida(solucion):  # Convierte la lista de símbolos en un string evaluable.
    return ''.join(solucion) # join() toma todos los elementos de la lista expresion y los une en una sola cadena.


def backtracking(solucion=[], etapa=0): # Genera todas las expresiones posibles de la forma: dig op dig op dig op dig op dig

    if len(solucion) == 0:
        solucion = [0 for _ in range(9)]  # 5 dígitos + 4 operadores
 
    if etapa % 2 == 0:    # Etapa es par, corresponde dígito
        candidatos = DIGITOS
    else:
        candidatos = OPERADORES

    for candidato in candidatos:
        solucion[etapa] = candidato

        if es_prometedora(solucion, etapa): # Verifica si la solución parcial es válida
            if etapa == 8: # Etapa final
                expr = expresion_valida(solucion)
                try:
                    valor = eval(expr)
                    if abs(valor - round(valor)) < 1e-9: # Verifica si el resultado es un valor entero
                        valor_int = int(round(valor))
                        enteros.add(valor_int) 
                        if valor_int not in resultados:
                            resultados[valor_int] = expr
                except ZeroDivisionError:
                    pass
            else:
                backtracking(solucion, etapa + 1) # Llamada recursiva para la siguiente etapa

        solucion[etapa] = 0


backtracking()

# Visualización de resultados
print("Cantidad de enteros:", len(enteros))
print("Mínimo:", min(enteros))
print("Máximo:", max(enteros))

faltantes = [x for x in range(min(enteros), max(enteros) + 1) if x not in enteros]

print("¿Están todos los enteros entre mínimo y máximo?", len(faltantes) == 0)
print("Algunos faltantes:", faltantes[:10])

print("\nEjemplos:")
for v in sorted(enteros)[:10]:
    print(v, "=>", resultados[v])
```

La técnica de vuelta atras es una construcción sistemática y por etapas de todas las soluciones posibles que pueden representarse mediante una tupla $(x_1, x_2, \dots, x_n)$, en la que cada componente $x_i$ puede explorarse en la **etapa $i$-ésima**. 

A través de un árbol de expansión se modela todo el espacio de soluciones, donde cada nodo representa un valor diferente para cada elemento $x_i$.

En el problema analizando el número de expresiones posibles crece rapidamente al aumentar la cantidad de dígitos y operadores. En este sentido, puede analizarse el crecimiento del espacio de búsqueda y como ello impacta en los tiempos de ejecución de la técnica de vuelta atrás.

Una mejora a considerar es incorporar estrategias de poda más eficientes, que permitan descartar ramas del árbol de búsqueda antes de completarlas. 

Asimismo, al aumentar el tamaño del problema (más dígitos u operadores), el algoritmo de backtracking puede volverse ineficiente. En estos casos, podrían evaluarse alternativas como: programación dinámica, si se identifican subproblemas repetidos y técnicas metaheurísticas, para explorar el espacio de soluciones de manera aproximada, o paralelización del algoritmo para distribuir la búsqueda.
